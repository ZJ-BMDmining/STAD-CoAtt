```python
sp_filtered = sp[~sp.index.isin(['1-Sep', '10-Sep', '11-Sep', '14-Sep', '15-Sep', '2-Mar', '2-Sep', '3-Sep', '4-Sep', '5-Sep', '6-Sep', '7-Sep', '8-Sep', '9-Sep', 'NRD1', 'SRPR'])]
sp_filtered
```

```python
import os
file = os.listdir("./GEO_data_5576")
# ROSMAP_file = [item for item in file if not item.startswith('G')]
# ROSMAP_file = [item for item in ROSMAP_file if not item.startswith('.')]
# ROSMAP_file
GEO_file = [item for item in file if item.startswith('G')]
GEO_file
```

```python
# 定义数据
# 这里需要提供图嵌入、48个病人矩阵、病人标签等数据
# graph_embedding = sp_filtered.values
# patient_matrix = []
# for i in ROSMAP_file:
#     df = pd.read_csv("./ROSMAP_data_5576/"+i,header = 0,index_col=0)
#     a = np.concatenate((df.T.values,graph_embedding),axis=1)
#     patient_matrix.append(a)
# patient_matrix = np.array(patient_matrix)

graph_embedding = sp_filtered.values
patient_matrix = []
for i in GEO_file:
    df = pd.read_csv("./GEO_data_5576/"+i,header = 0,index_col=0)
    a = np.concatenate((df.T.values,graph_embedding),axis=1)
    patient_matrix.append(a)
patient_matrix = np.array(patient_matrix)
```

```python
# labels = pd.read_csv("./label_ROSMAP.csv",header=0,index_col=0).T
# f = []
# for i in ROSMAP_file:
#     f.append(int(i.split(".")[0]))
# labels = labels[f]
# labels = labels.T
# patient_labels = labels.values[:,0]

labels = pd.read_csv("./label_GEO.csv",header=0)
f = []
for i in GEO_file:
    f.append(i.split("_")[0])
labels = labels.set_index('Unnamed: 0').loc[f].reset_index()
l = np.array(labels['Braak']).astype(np.int32)
for i in range(len(l)):
    if l[i]==2:
        l[i]=1
for i in range(len(l)):
    if l[i]==6:
        l[i]=2
l
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score
# label = np.array(labels["CERAD"])
label = l
X = patient_matrix.reshape((patient_matrix.shape[0], patient_matrix.shape[1], patient_matrix.shape[2], 1))
X_train, X_val, y_train, y_val = train_test_split(X, label, test_size=0.2, random_state=42)
y_train = y_train.astype(np.int32)
y_val = y_val.astype(np.int32)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
# with tf.device("/gpu:0"):
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(X.shape[1], X.shape[2], 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(256, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(3, activation='softmax'))
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
# from tensorflow.keras.callbacks import ModelCheckpoint

# checkpoint = ModelCheckpoint('best_model_spatialfusion_Braak.h5', monitor='val_loss', save_best_only=True)

with tf.device("/cpu:0"):
    # history = model.fit(X_train, y_train, epochs=1, batch_size=4, validation_data=(X_val, y_val), callbacks=[checkpoint])
    history = model.fit(X_train, y_train, epochs=1, batch_size=2, validation_data=(X_val, y_val))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
# best_model = tf.keras.models.load_model('best_model_spatialfusion_Braak.h5')
with tf.device("/cpu:0"):
    # predictions = best_model.predict(X)
    predictions = model.predict(X)
    
predicted_labels = np.argmax(predictions, axis=1)

# 计算准确率
accuracy = accuracy_score(label, predicted_labels)
print(f'Test Accuracy: {accuracy}')

# 计算精确度
precision = precision_score(label, predicted_labels, average='weighted')
print(f'Precision: {precision}')

# 计算召回率
recall = recall_score(label, predicted_labels, average='macro')
print(f'Recall: {recall}')

# 计算F1分数
f1 = f1_score(label, predicted_labels, average='weighted')
print(f'F1 Score: {f1}')
```

```python
predicted_labels = np.array([pred for pred in predicted_labels.tolist()])
df = pd.DataFrame({'pred': predicted_labels, 'true': label})
df
```
